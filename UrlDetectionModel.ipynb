4번 특징 domain, path, patameter, TLD 길이가 길다

```python
import pandas as pd
from urllib.parse import urlparse, parse_qs

df = pd.read_csv('/content/dataset.csv')
```

```python
url_http = df[df['url'].str.contains('http')]

def extract_url_info(url):
    parsed_url = urlparse(url)

    # Set netloc format netloc 형식 구성
    if len(parsed_url.netloc) == 0:
        parsed_url = urlparse("//"+url)

    # Extract domain length domain 길이 추출
    domain = parsed_url.netloc
    domain_length = len(domain)

    # Extract path length path 길이 추출
    path = parsed_url.path
    path_length = len(path)

    # Extract parameter length parameter 길이 추출
    query = parsed_url.query
    parameters = parse_qs(query)
    parameter_length = sum(len(v) for v in parameters.values())

    # Extract TLD (Top-Level Domain) length TLD 길이 추출
    tld = parsed_url.netloc.split('.')[-1]
    tld_length = len(tld)

    return domain_length, path_length, parameter_length, tld_length
```

```python
url_http_copy = url_http.copy()
# url_http_copy = df.copy()
url_http_copy[['domain_length', 'path_length', 'parameter_length', 'tld_length']] = url_http_copy['url'].apply(lambda x: pd.Series(extract_url_info(x)))
```

1번특징 생성, 갱신일자 확인, 만료일자 서비스 기간 짧음




```python
#도메인 유효 기간을 계산하는 함수
#일부 변형해서 갱신일자 가져올 수 있을 것 같음
import whois, sys
import urllib.parse
from datetime import datetime

# domain 유효 기간 계산
def get_total_date(url) :
    domain = whois.whois(url) #도메인 추출
    if type(domain.expiration_date) is list : #도메인에서 만료일 추출, 리스트 형태면 첫번째 값 가져옴
        expiration_date = domain.expiration_date[0]
    else :
        expiration_date = domain.expiration_date

    if type(domain.updated_date) is list : #도메인에서 갱신일 추출, 리스트 형태면 첫번째 값 가져옴
        updated_date = domain.updated_date[0]
    else :
        updated_date = domain.updated_date
    if updated_date is None:#결과 없을 시 none 리턴
        return None
    if expiration_date is None:
        return None

    #도메인 유효 기간 계산(만료일-갱신일)
    total_date = (expiration_date - updated_date).days

    return total_date
```

```python
#위에 정의한 함수 사용해서 도메인 유효기간 배열 생성

domainDate = []
dataset = output

for i, item in dataset.iterrows():
    try:
        date = get_total_date(item['url'])
        if date is None: #함수 결과값이 없을 경우
            domainDate.append(np.nan) #Nan 넣어둠
        else:
            domainDate.append(date)
    except: #양식이 없는 양식일 경우(함수에러가 뜸)
        domainDate.append(np.nan)
        continue

dataset['domainDate'] = domainDate
dataset
# domainDate = pd.DataFrame(domainDate, columns=['domainDate'])

# #원활한 저장 위해 인덱스 초기화
# dataset = dataset.reset_index()
# domainDate = domainDate.reset_index()

# #데이터프레임 병합(index가 여러개 생기는 것 같은데 해결해야할수도)
# dataset = pd.concat([dataset, domainDate], axis=1)
```

```python
#url에서 domain 추출

import csv
from urllib.parse import urlparse

input_file = 'unique_url.csv'
output_file = 'domain.csv'

output_header = ['URL', 'Domain']

# 도메인 추출 함수 정의
def extract_domain(url):
    try:
        parsed_url = urlparse(url)
        domain = parsed_url.netloc
        return domain
    except Exception as e:
        print(f"Error: {e}")
        return None

with open(input_file, 'r') as file:

    reader = csv.reader(file)

    domains = []

    # 행 단위로 데이터 처리
    for row in reader:

        url = row[0]

        # 도메인 추출
        domain = extract_domain(url)

        domains.append((url, domain))

with open(output_file, 'w', newline='') as output:
    writer = csv.writer(output)

    # 헤더 쓰기
    writer.writerow(output_header)

    writer.writerows(domains)
```

```python
#domain에서 ip 추출

import csv
import dns.resolver

def ip_from_domain(domain):
    try:
        r = dns.resolver.Resolver()
        r.nameserver = ['8.8.8.8']
        ip = dns.resolver.resolve(domain, 'A')
        if ip:
            return str(ip[0])
    except Exception as e:
        print(f"Error: {e}")
        return None

csv_file = "domain.csv"

output_csv_file = "ip.csv"

parsed_ip = []

with open(csv_file, "r", encoding='utf-8') as file:
    reader = csv.DictReader(file)

    with open(output_csv_file, "w", newline="", encoding='utf-8') as output_file:
        writer = csv.writer(output_file)
        writer.writerow(["ip"])

        for row in reader:
            domain = row['domain']
            ip = ip_from_domain(domain)
            parsed_ip.append(ip)
            writer.writerow([ip])
```

```python
#ip에서 netname, asn 추출

from ipwhois import IPWhois
import pandas as pd
import csv
from concurrent.futures import ProcessPoolExecutor
import os

def get_ip_information(ip):
    try:
        ip_info = {}
        w = IPWhois(ip)
        result = w.lookup_rdap()

        ip_info['IP'] = ip
        ip_info['NetName'] = result.get('asn_description', '') #netname 추출
        ip_info['ASN'] = result.get('asn', '') #asn 추출
        return ip_info
    except Exception as e:
        print(f"Error: {e}")
        return None

ip_dataset_file = "ip.csv"
output_csv_file = "ip_info.csv"

df = pd.read_csv(ip_dataset_file)

df['IP'] = df['IP'].astype(str)

# 중복된 IP 주소 제거
unique_ips = df['IP'].unique()

max_workers = os.cpu_count()

# 멀티프로세싱으로 IP 정보 추출
ip_information = []
try:
    with ProcessPoolExecutor(max_workers=max_workers) as executor:
        results = executor.map(get_ip_information, unique_ips)
        for result in results:
            if result is not None:
                ip_information.append(result)
except Exception as e:
    print(f"Error: {e}")

with open(output_csv_file, "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=['IP', 'NetName', 'ASN'])
    writer.writeheader()
    writer.writerows(ip_information)
```

```python
#netname에서 중복된 값 제거

import csv

input_file = 'netname.csv'
output_file = 'unique_netname.csv'

# 중복된 값을 저장할 집합(set) 생성
unique_values = set()

with open(input_file, 'r') as file:

    reader = csv.reader(file)

    unique_rows = []

    # 행 단위로 데이터 처리
    for row in reader:
        # 행에 공백이 있는 경우 무시
        if not any(row):
            continue

        # 현재 행의 값을 문자열로 변환하여 중복 여부 확인
        row_string = ','.join(row)

        # 중복된 값이 이미 존재하는 경우 해당 행을 건너뛰고 다음 행으로 진행
        if row_string in unique_values:
            continue

        # 중복된 값이 존재하지 않는 경우 집합과 리스트에 행 추가
        unique_values.add(row_string)
        unique_rows.append(row)

with open(output_file, 'w', newline='') as output:
    writer = csv.writer(output)
    writer.writerows(unique_rows)
```

```python
import pandas as pd
import socket
import dns.resolver
from ipwhois import IPWhois
from sklearn.ensemble import GradientBoostingClassifier
from urllib.parse import urlparse
from sklearn.metrics import accuracy_score, precision_score, recall_score, fbeta_score, roc_auc_score

# 전체 데이터셋 로드
url_dataset = pd.read_csv("unique_dataset.csv")

# 악성 URL의 netname과 asn을 저장한 CSV 파일 로드
unique_netname = pd.read_csv("unique_netname.csv")
unique_asn = pd.read_csv("unique_asn.csv")

# URL에서 도메인을 추출하는 함수
def extract_domain(url):
    try:
        parsed_url = urlparse(url)
        if len(parsed_url.netloc) == 0:
            parsed_url = urlparse("//"+url)
        domain = parsed_url.netloc
        return domain
    except Exception as e:
        print(f"Error: {e}")
        return None

# 도메인에서 IP를 추출하는 함수
def extract_ip(domain):
    try:
        r = dns.resolver.Resolver()
        r.nameserver = ['8.8.8.8']
        ip = dns.resolver.resolve(domain, 'A')
        if ip:
            return ip
    except Exception as e:
        print(f"Error: {e}")
        return None

# IP 주소로부터 netname 추출하는 함수
def extract_netname(ip):
    try:
        hostname = socket.gethostbyaddr(ip)[0]
        w = IPWhois(hostname)
        result = w.lookup_rdap(depth=1)
        netname = result['asn_description']
        return netname
    except Exception as e:
        return None

# IP 주소로부터 asn 추출하는 함수
def extract_asn(ip):
    try:
        hostname = socket.gethostbyaddr(ip)[0]
        w = IPWhois(hostname)
        result = w.lookup_rdap(depth=1)
        asn = result['asn']
        return asn
    except Exception as e:
        return None

url_dataset['ip'] = url_dataset['url'].apply(extract_ip)
url_dataset['netname'] = url_dataset['ip'].apply(extract_netname)
url_dataset['asn'] = url_dataset['ip'].apply(extract_asn)

url_dataset['is_malicious'] = url_dataset.apply(lambda row: (row['netname'] in unique_netname['netname'].values) and (row['asn'] in unique_asn['asn'].values), axis=1)

# URL 길이를 기반으로 특징 벡터 생성
url_dataset['domain_length'] = url_dataset['domain'].apply(len)
url_dataset['path_length'] = url_dataset['path'].apply(len)
url_dataset['tld_length'] = url_dataset['tld'].apply(len)

# URL 도메인 형식을 확인하여 특징 벡터 생성
url_dataset['is_ipv4'] = url_dataset['domain'].apply(lambda x: 1 if x.count('.') == 3 else 0)
url_dataset['is_ipv6'] = url_dataset['domain'].apply(lambda x: 1 if x.count(':') >= 2 else 0)

# 특징 벡터와 레이블 분리
features = url_dataset[['domain_length', 'path_length', 'tld_length', 'is_ipv4', 'is_ipv6']]
labels = url_dataset['is_malicious']

# 그래디언트 부스팅 알고리즘을 사용한 모델 학습
model = GradientBoostingClassifier()
model.fit(features, labels)

# 평가 데이터셋 로드
evaluation_dataset = pd.read_csv("unique_dataset.csv")

# 평가 데이터셋에 대한 특징 추출
evaluation_dataset['domain_length'] = evaluation_dataset['domain'].apply(len)
evaluation_dataset['path_length'] = evaluation_dataset['path'].apply(len)
evaluation_dataset['tld_length'] = evaluation_dataset['tld'].apply(len)
evaluation_dataset['is_ipv4'] = evaluation_dataset['domain'].apply(lambda x: 1 if x.count('.') == 3 else 0)
evaluation_dataset['is_ipv6'] = evaluation_dataset['domain'].apply(lambda x: 1 if x.count(':') >= 2 else 0)

# 평가 데이터셋에 대한 예측 수행
evaluation_features = evaluation_dataset[['domain_length', 'path_length', 'tld_length', 'is_ipv4', 'is_ipv6']]
predictions = model.predict(evaluation_features)

# 평가 지표 계산
accuracy = accuracy_score(evaluation_dataset['type'], predictions)
precision = precision_score(evaluation_dataset['type'], predictions)
recall = recall_score(evaluation_dataset['type'], predictions)
beta = 1
f1_beta = (1 + beta ** 2) * (precision * recall) / ((beta ** 2 * precision) + recall)
specificity = recall_score(evaluation_dataset['type'], predictions, pos_label='good')
false_positive_rate = 1 - specificity
miss_rate = 1 - recall
roc_auc = roc_auc_score(evaluation_dataset['type'], predictions)

# 평가 지표 출력
print("정확도:", accuracy)
print("정밀도:", precision)
print("재현율:", recall)
print("F1-score:", f1_beta)
print("특이도:", specificity)
print("오탐률:", false_positive_rate)
print("미탐률:", miss_rate)
print("ROC AUC:", roc_auc)
```
